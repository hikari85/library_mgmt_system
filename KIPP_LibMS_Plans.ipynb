## Essential Triggers

1. Trigger when Issuing a Book - Checks the no_of_copies available
```
if no_of_copies>0:
    issues the book
    no_of_copies = no_of_copies - 1
elif no_of_copies=0:
    then the book cannot be issued
```

2. Trigger when Issuing a Book -

Checks if the User has any dues
Given the User_ID(USN or Faculty_ID, etc) checks the 'is_returned' column from 'Registry_for_Issuing' Table and also compares today's date with due_date
```
if is_returned==True:
    User is eligible for the new Issue
elif is_returned==False and due_date>today_date:
    User is eligible for the new Issue, with a caution of item(s) to be returned
else:
    User has due(s), collect item(s) back before issuing new item(s)
```

3. Trigger when a Book is Returned - Changes no_of_copies in the Books table

Whenever a book is returned, is_returned becomes True and then the no_of_copies for the corresponding Asset_ID gets added by one
```
after is_returned==True:
    no_of_copies = no_of_copies + 1
```
